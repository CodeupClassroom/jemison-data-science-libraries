```python
import pandas as pd
from pydataset import data
```

```python
df = data('tips')
df.head()
```

```python
pd.crosstab([df.day, df.smoker], df.time)
```

```python
pd.DataFrame(df.groupby(["day", "smoker", "time"]).time.count())
```

```python
# Pivot table's computation defaults to the mean
df.pivot_table(index="day", columns="time", values="tip")
```

```python
df.pivot_table(index="day", columns="time", values="tip", aggfunc="median")
```

```python
df[(df.time == "Lunch") & (df.day == "Sun")]
```

```python
df[(df.time == "Lunch") & (df.day == "Sun")].mean()
```

```python
df.pivot_table(index="day", columns="time", values="tip", aggfunc=["mean", "median"])
```

```python
pd.DataFrame(df.groupby(["time"]).tip.mean())
```

```python
pd.DataFrame(df.groupby(["time"]).tip.mean()).index
```

```python
# Grouping by multiple columns creates a multi-level index
pd.DataFrame(df.groupby(["time", "day"]).tip.mean())
```

```python
pd.DataFrame(df.groupby(["time", "day"]).tip.mean()).index
```

```python
# .reset_index on a multi-level index turns the multi-level indexes into new columns
pd.DataFrame(df.groupby(["time", "day"]).tip.mean()).reset_index()
```
